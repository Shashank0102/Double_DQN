```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from agent import Agent
```

```python
game_name = 'AmidarNoFrameskip-v4' # 'Breakout-v0' # 'VideoPinball-v0'
device = 'cuda:0'
chkpnt_name = 'AmidarNoFrameskip-v4_5000' # 'breakoutv4_100_test'
pretrained = 'AmidarNoFrameskip-v4_5000' # 'breakoutv4_100_test'

agent = Agent(game_name, device, chkpnt_name, pretrained_name=pretrained, verbosity=1)

plt.imshow(agent.sanity_check_screen())
plt.plot()
```

```python
metrics = agent.train()
```

```python
if 'metrics' not in locals():
    import pickle
    with open('./models/PongNoFrameskip-v0_10000.metrics', 'rb') as metrics_file:
        metrics = pickle.load(metrics_file)

rewards = metrics['rewards']
losses = metrics['losses']
steps = metrics['steps']
cum_steps = metrics['cum_steps']

assert len(rewards) == len(losses) == len(steps)

episodes_info = list(zip(rewards, losses, steps, cum_steps))
for i, (reward, loss, step, cum_step) in enumerate(episodes_info):
    print('Episode %d :' % i)
    print('Step - ', step)
    print('Cumulative steps - ', cum_step)
    print('Reward - ', reward)
    print('Loss - ', loss)
    print('\n')
```

```python
plt.plot(range(len(losses)), losses)
plt.show()
```

```python
plt.plot(range(len(rewards)), rewards)
plt.show()
```

```python
moving_avg_window = 100
averaged_rewards = np.convolve(rewards, np.ones(moving_avg_window), 'valid') / moving_avg_window

plt.plot(range(len(averaged_rewards)), averaged_rewards)
plt.show()
```

```python
moving_avg_window = 10
eval_reward = 0
for _ in range(moving_avg_window):
    _R = agent.evaluate(visualise=False)
    print(_R)
    eval_reward += _R
eval_reward /= moving_avg_window

print(eval_reward)
```
